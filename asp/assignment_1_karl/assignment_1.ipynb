```python
import os
from clingo.control import Control

env_encoding_file = "env.lp"
```

```python
with open(env_encoding_file, "r") as f:
    env_encoding = f.read()
ctl = Control()
ctl.add(env_encoding)
ctl.ground()
grid = {}
with ctl.solve(yield_ = True) as handle:
    for model in handle:
        for symbol in model.symbols(shown = True):
            if symbol.name == "cell":
                # print(f"symbol: {symbol}")
                x = symbol.arguments[0].arguments[0].number
                y = symbol.arguments[0].arguments[1].number
                rail_type = symbol.arguments[1].number
                if x not in grid:
                    grid[x] = {}
                if grid[x].get(y) is not None:
                    print(f"Warning: overwriting cell at x: {x}, y: {y} from {grid[x][y]} to {rail_type}")
                grid[x][y] = rail_type
                # print(f"x: {x}, y: {y}, rail_type: {rail_type}")

# sort grid by x and y
sorted_grid = {}
for x in sorted(grid.keys()):
    sorted_grid[x] = {}
    for y in sorted(grid[x].keys()):
        sorted_grid[x][y] = grid[x][y]
for x in sorted_grid:
    print(f"x: {x}, length: {len(sorted_grid[x])}")
    for y in sorted_grid[x]:
        print(f"  y: {y}, rail_type: {sorted_grid[x][y]}")
```
